## Visualisasi Data beserta Distribusinya

```python
sns.set(style="whitegrid")
```

```python
# Visualisasi distribusi 'type'
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution of Type (Movies vs TV Shows)', fontsize=14)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

Distribusi type disini memaparkan jumlah dari konten yaitu Movie dan TV Show. Selain itu, distribusi ini membantu untuk memahami proporsi konten berdasarkan jenisnya. Termasuk Distribusi Kategori (Diskrit). Lebih jelasnya yaitu Distribusi Nominal, karena kategori "Movie" dan "TV Show" tidak memiliki urutan atau hierarki, hanya menunjukkan klasifikasi data yang berbeda.

```python
# Visualisasi distribusi 'release_year'
plt.figure(figsize=(10, 6))
sns.histplot(df['release_year'], bins=30, kde=True, color='purple')
plt.title('Distribution of Release Year', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

Pada distribusi release_year disini memaparkan persebaran konten di platform Disney+ berdasarkan tahun rilis. Histogram yang muncul juga menggambarkan banyaknya judul yang rilis tiap tahunnya, dengan garis kepadatan (KDE) yang menunjukkan distribusi konten secara lebih halus. Termasuk Distribusi Kontinu. Distribusi Kontinunya yaitu Distribusi Miring ke Kanan terjadi ketika sebagian besar data berada di sisi kiri distribusi (nilai-nilai yang lebih rendah), dan ada ekor panjang ke sisi kanan (nilai yang lebih tinggi), seperti yang terlihat pada data tahun rilis konten Disney+ yang terkonsentrasi setelah tahun 2000, dengan sedikit konten di tahun-tahun sebelumnya.

```python
# Visualisasi distribusi 'rating'
plt.figure(figsize=(10, 6))
sns.countplot(y='rating', data=df, palette='cool', order=df['rating'].value_counts().index)
plt.title('Distribution of Content Rating', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.show()
```

Distribusi diatas yaitu distribusi rating menunjukkan jumlah dari konten yang berdasarkan rating seperti TV-PG, PG, dan lainnya. Bar plot horizontal digunakan untuk menunjukkan jumlah judul di setiap kategori rating. Termasuk Distribusi Kategori (Diskrit) yaitu termasuk dalam Distribusi Frekuensi. Jenis distribusi ini berguna untuk memahami frekuensi berbagai kategori dalam dataset.

```python
# Visualisasi distribusi 'country'
plt.figure(figsize=(10, 8))
top_countries = df['country'].value_counts().head(10)  # Menampilkan 10 negara teratas
sns.barplot(y=top_countries.index, x=top_countries.values, palette="Blues_d")
plt.title('Top 10 Countries by Number of Titles', fontsize=14)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
```

Distribusi ini disebut Distribusi Frekuensi Kategori, yang mana dapat dilihat seberapa sering suatu negara tampil sebagai tempat produksi di platform Disney+. ## Pemusatan dan Penyebaran Data beserta Analisis Skewness 

```python
# 1. Statistik Pemusatan dan Penyebaran

def central_tendency_and_dispersion(column_name):
    """Menghitung dan menampilkan statistik pemusatan dan penyebaran untuk kolom tertentu."""
    stats = df[column_name].describe()
    mode = df[column_name].mode()[0]  # Nilai yang paling sering muncul
    
    print(f"Statistik Pemusatan dan Penyebaran untuk '{column_name}':")
    print(stats)
    print(f"\nMode (Nilai yang Paling Sering Muncul): {mode}")
    
    # Hanya menghitung skewness untuk kolom numerik
    if pd.api.types.is_numeric_dtype(df[column_name]):
        skewness = df[column_name].skew()  # Kemiringan distribusi (skewness)
        print(f"Skewness (Kemiringan): {skewness}")
        
        # Analisis skewness
        if skewness > 0:
            print(f"Distribusi '{column_name}' miring ke kanan (positif skewness).")
        elif skewness < 0:
            print(f"Distribusi '{column_name}' miring ke kiri (negatif skewness).")
        else:
            print(f"Distribusi '{column_name}' simetris (skewness mendekati 0).")
    else:
        print(f"Kolom '{column_name}' adalah kategorikal, sehingga skewness tidak dihitung.")
    
    print("\n" + "-"*50 + "\n")


# 2. Hitung untuk 9 distribusi di atas:

# 1. Distribusi 'type' - Kategori (Diskrit)
central_tendency_and_dispersion('type')

# 2. Distribusi 'release_year' - Numerik (Kontinu)
central_tendency_and_dispersion('release_year')

# 3. Distribusi 'rating' - Kategori (Diskrit)
central_tendency_and_dispersion('rating')

# 4. Distribusi 'country' - Kategori (Diskrit)
central_tendency_and_dispersion('country')

# 5. Distribusi 'duration' - Numerik (Kontinu, tergantung format durasi)
df['duration_numerical'] = df['duration'].str.extract('(\d+)').astype(float)  # Ekstrak angka dari durasi
central_tendency_and_dispersion('duration_numerical')

# 6. Distribusi 'director' - Kategori (Diskrit)
central_tendency_and_dispersion('director')

# 7. Distribusi 'listed_in' (Genre) - Kategori (Diskrit)
central_tendency_and_dispersion('listed_in')

# 8. Distribusi 'date_added' - Tanggal (Kontinu)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
central_tendency_and_dispersion('date_added')

# 9. Distribusi Berdasarkan Jumlah Aktor ('cast') - Kategori (Diskrit)
central_tendency_and_dispersion('cast')
```

Kode diatas berguna untuk menggabungkan analisis statistik untuk data kategorikal dan numerik dari dataset. Skewness hanya dihitung untuk kolom numerik, sedangkan statistik deskriptif umum diterapkan untuk kolom kategorikal. Analisis seperti ini berguna untuk memahami pemusatan (mean, median, mode) dan penyebaran (standard deviation, min, max) dari berbagai atribut konten di Disney+.
